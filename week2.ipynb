## Q6. 데이터 전처리 및 집계(10점)

```python
''' 데이터 확인용 셀입니다. '''

visit_df = pd.read_csv('/content/drive/MyDrive/LLM_과제/data/eda_visit.csv')
display(visit_df.head())
```

```python
''' 정답 코드를 작성해주세요 '''

''' 참고자료 : 3-10 values & NA .ipynb '''

# 결측치 제거 수행
# 옵션 - any
```

```python
visit_df_cleaned = visit_df.dropna(how='any')
visit_df_cleaned
```

```python
''' 정답 코드를 작성해주세요 '''

''' 참고자료 : 3-07 groupby.ipynb '''

# 결과

# campaign, source 별로 user 숫자 구하기
# nunique()를 집계함수로 적용
# rename을 통해 집계 컬럼이름은 user_num으로 변경
```

```python
grouped = visit_df_cleaned.groupby(['campaign', 'source'])['user_id'].nunique().reset_index()
grouped.rename(columns={'user_id': 'user_num'}, inplace=True)
grouped
```
